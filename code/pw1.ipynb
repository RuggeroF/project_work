```python
import pandas as pd
```

```python
arr = pd.read_csv("tin00174_linear.csv")
```

```python
arr
```

```python
arr = arr[["geo", "TIME_PERIOD", "c_resid", "OBS_VALUE"]]
arr.columns = ["country", "year", "resid", "arrivals"]
arr.sort_values(by=['resid', 'country', 'year'], inplace=True)
arr.reset_index(drop=True, inplace=True)
```

```python
print(arr.to_string())
```

```python
arr.drop(arr[arr['country'] == "European Union - 27 countries (from 2020)"].index, inplace=True)
arr.reset_index(drop=True, inplace=True)
```

```python
print(arr.to_string())
```

```python
arr.dtypes
```

```python
arrivals_wide = arr.pivot_table(
    index=["country", "year"],
    columns="resid",
    values="arrivals",
    aggfunc="sum"
).reset_index()
```

```python
arrivals_wide = arrivals_wide.rename(columns={
    "Domestic country": "arrivals_dom",
    "Foreign country": "arrivals_for",
    "Total": "arrivals_tot"
})
```

```python
arrivals_wide = arrivals_wide[[
    "country",
    "year",
    "arrivals_tot",
    "arrivals_dom",
    "arrivals_for"
]]
```

```python
arrivals_wide = arrivals_wide.sort_values(["country", "year"]).reset_index(drop=True)
```

```python
cols = ["arrivals_tot", "arrivals_dom", "arrivals_for"]

arrivals_wide[cols] = arrivals_wide[cols].astype("Int64")
```

```python
arrivals_wide.columns.name = None
```

```python
print(arrivals_wide.to_string())
```
