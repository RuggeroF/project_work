```python
import pandas as pd
```

```python
arr = pd.read_csv("../res/tin00174_linear.csv")
```

```python
arr
```

```python
arr = arr[["geo", "TIME_PERIOD", "c_resid", "OBS_VALUE"]]
arr.columns = ["country", "year", "resid", "arrivals"]
arr.sort_values(by=['resid', 'country', 'year'], inplace=True)
arr.reset_index(drop=True, inplace=True)
```

```python
print(arr.to_string())
```

```python
arr.drop(arr[arr['country'] == "European Union - 27 countries (from 2020)"].index, inplace=True)
arr.reset_index(drop=True, inplace=True)
```

```python
print(arr.to_string())
```

```python
arr.dtypes
```

```python
arrivals_wide = arr.pivot_table(
    index=["country", "year"],
    columns="resid",
    values="arrivals",
    aggfunc="sum"
).reset_index()
```

```python
arrivals_wide = arrivals_wide.rename(columns={
    "Domestic country": "arrivals_dom",
    "Foreign country": "arrivals_for",
    "Total": "arrivals_tot"
})
```

```python
arrivals_wide = arrivals_wide[[
    "country",
    "year",
    "arrivals_dom",
    "arrivals_for",
    "arrivals_tot"
]]
```

```python
arrivals_wide = arrivals_wide.sort_values(["country", "year"]).reset_index(drop=True)
```

```python
print(arrivals_wide.to_string())
```

```python
cols = ["arrivals_tot", "arrivals_dom", "arrivals_for"]
arrivals_wide[cols] = arrivals_wide[cols].astype("Int64")
arrivals_wide.columns.name = None
```

```python
print(arrivals_wide.to_string())
```

```python
arrivals_wide.dtypes
```

```python
# Carica file
nights = pd.read_csv("../res/tin00175_linear.csv")
```

```python
nights
```

```python
# Tieni solo le colonne utili
nights = nights[["geo", "TIME_PERIOD", "c_resid", "OBS_VALUE"]]

# Rinomina
nights = nights.rename(columns={
    "geo": "country",
    "TIME_PERIOD": "year",
    "c_resid": "resid",
    "OBS_VALUE": "nights"
})

# Ordina
nights = nights.sort_values(
    by=["resid", "country", "year"]
).reset_index(drop=True)
```

```python
print(nights.to_string())
```

```python
# Elimina aggregato UE
nights = nights[
    nights["country"] != "European Union - 27 countries (from 2020)"
].reset_index(drop=True)
```

```python
print(nights.to_string())
```

```python
nights.dtypes
```

```python
# Long -> Wide
nights_wide = nights.pivot_table(
    index=["country", "year"],
    columns="resid",
    values="nights",
    aggfunc="sum"
).reset_index()
```

```python
# Rinomina colonne
nights_wide = nights_wide.rename(columns={
    "Total": "nights_tot",
    "Domestic country": "nights_dom",
    "Foreign country": "nights_for"
})
```

```python
# Rimuove il nome "resid" sopra le colonne
nights_wide.columns.name = None

nights_wide
```

```python
# Interi
cols = ["nights_tot", "nights_dom", "nights_for"]
nights_wide[cols] = nights_wide[cols].astype("Int64")
```

```python
# Controlli
print(nights_wide.head())
print(nights_wide.shape)
print(nights_wide.isna().sum())

print(nights_wide.to_string())
```

```python
tourism = arrivals_wide.merge(
nights_wide,
on=["country", "year"],
how="outer"
)

tourism
```

```python
purpose = pd.read_csv("../res/tin00191_linear.csv")
```

```python
# Tieni solo le colonne utili
purpose = purpose[["purpose", "geo", "TIME_PERIOD", "OBS_VALUE"]]

# Rinomina
purpose = purpose.rename(columns={
    "purpose": "purpose",
    "geo": "country",
    "TIME_PERIOD": "year",
    "OBS_VALUE": "nights_purpose"
})

# Ordina
purpose = purpose.sort_values(
    by=["country", "year"]
).reset_index(drop=True)

purpose = purpose[
    purpose["country"] != "European Union - 27 countries (from 2020)"
].reset_index(drop=True)


print(purpose.to_string())
purpose.dtypes
```

```python
cols = ["nights_purpose"]
purpose[cols] = purpose[cols].astype("Int64")

print(purpose.to_string())
purpose.dtypes
```

```python
purpose_wide = purpose.pivot_table(
    index=["country", "year"],
    columns="purpose",
    values="nights_purpose",
    aggfunc="sum"
).reset_index()


purpose_wide = purpose_wide.rename(columns={
    "Holidays, leisure and recreation": "purpose_leisure",
    "Professional, business": "purpose_work",
    "Visits to friends and relatives": "purpose_visit",
    "Personal reasons": "purpose_personal",
    "Other personal reasons": "purpose_other",
    "Total": "purpose_tot"
})

purpose_wide = purpose_wide[[
    "country",
    "year",
    "purpose_leisure",
    "purpose_work",
    "purpose_visit",
    "purpose_personal",
    "purpose_other",
    "purpose_tot"
]]

print(purpose_wide.to_string())
```

```python
tourism = tourism.merge(
purpose_wide,
on=["country", "year"],
how="outer"
)

tourism
```

```python
tourism.to_csv('../res/report_tourism.csv', sep=',', encoding='utf-8')
```
